```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('/Users/sky/class_python/6.Recommendation/jobkorea_saramin_recruit.csv', index_col=0)
# 인덱스 중복되는게 나와서 초기화하고 시작. ㅇㅅㅇ
df.reset_index(inplace=True)
df.head(3)
```

```python
company = df['company'].tolist()
```

```python
category = df['category']
category = category.to_list()
type(category)
```

```python
#Counter Vectorizer 형태로 변형하깅
count =  CountVectorizer(max_features=100)
a = count.fit_transform(category)
a
```

47645개 문장에 6444 단어가 이씀. -> 100개만 써보기로.

```python
a.toarray()
```

```python
# value 기준으로 정렬
sorted(count.vocabulary_, reverse=True)
```

```python
# matrix에 인덱스와 컬럼 값 넣기
import pandas as pd
count_df = pd.DataFrame(a.toarray(), columns= sorted(count.vocabulary_))
count_df.index = company
count_df
```

# 유사도 계산

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_table = cosine_similarity(count_df, count_df)
```

```python
#sim_table.shape
np.argsort(sim_table[0]) #기본 오름차순 정렬
```

```python
from sklearn.metrics.pairwise import euclidean_distances

#유클리디언, 맨해튼은 정규화가 필요함
def normalize(v):
    norm = np.sum(v)
    return v / norm

norm_df = normalize(count_df)

print(euclidean_distances(norm_df[0:1], norm_df[1:2]))
```

```python
from sklearn.metrics.pairwise import manhattan_distances
print(manhattan_distances(norm_df[0:1], norm_df[1:2]))
```

# TF-IDF

```python
tfidf =  TfidfVectorizer(max_features=100)
b = tfidf.fit_transform(category)
b
```

```python
tfidf_df = pd.DataFrame(b.toarray(), columns=sorted(tfidf.vocabulary_))
tfidf_df.index = company
tfidf_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_table2 = cosine_similarity(tfidf_df, tfidf_df)
np.argsort(sim_table2[0])
```

```python
from sklearn.metrics.pairwise import euclidean_distances

def normalize(v):
    norm = np.sum(v)
    return v / norm

norm_b = normalize(tfidf_df)

print(euclidean_distances(norm_b[0:1], norm_b[1:2]))
```

```python
from sklearn.metrics.pairwise import manhattan_distances

def normalize(v):
    norm = np.sum(v)
    return v / norm

norm_b = normalize(tfidf_df)

print(manhattan_distances(norm_b[0:1], norm_b[1:2]))
```

# Y9YA 유사도 계산은 했지만 해당 인덱스가 어떤 기업인지 모린다.

```python
# company와 index를 매핑할 딕셔너리 만들깅

company2idx = {}
for i, c in enumerate(df['company']): 
    company2idx[i] = c
    
# idx와 company를 매핑할 딕셔너리 만들깅

idx2company = {}
for i, c in company2idx.items(): idx2company[c] = i
```

```python
company2idx[0]
```

```python
idx2company
```

# 아프리카TV 채용정보랑 비슷한 기업 찾아보깅

```python
# 1. 아프리카티비 index 추출
idx = idx2company['㈜아프리카TV']
idx
```

```python
# 자기 자신을 제외한 기업들의 유사도 및 인덱스 추출하깅
sim_scores = [(i, c) for i, c in enumerate(sim_table2[idx]) if i != idx]
# 유사도 높은 순으로 정렬
sim_scores = sorted(sim_scores, key= lambda x: x[1], reverse=True)
sim_scores[0:10]
```

```python
# 인덱스를 기업명으로 바까주깅
sim_scores = [(company2idx[i], score) for i, score in sim_scores[0:10]]
sim_scores
```
