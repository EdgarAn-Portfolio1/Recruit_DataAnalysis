## degree 컬럼 전처리

```python
df['degree'].value_counts()
```

```python
df['degree'] = df['degree'].str.replace('고졸↑','고등학교 졸업')
df['degree'] = df['degree'].str.replace('초대졸↑','대학졸업(2,3년)')
df['degree'] = df['degree'].str.replace('대학(2,3년)↑','대학졸업(2,3년)')
df['degree'] = df['degree'].str.replace('대졸↑','대학교졸업(4년)')
df['degree'] = df['degree'].str.replace('대학교(4년)↑','대학교졸업(4년)')
df['degree'] = df['degree'].str.replace('석사↑','대학원 석사졸업')
df['degree'] = df['degree'].str.replace('박사','대학원 박사졸업')
df['degree'] = df['degree'].str.replace('박사↑','대학원 박사졸업')
df['degree'] = df['degree'].str.replace('학력무관','그 외')
df['degree'] = df['degree'].str.replace('-','그 외')
```

```python
df['degree'] = df['degree'].replace('대학(2,3년)↑','대학졸업(2,3년)')
df['degree'] = df['degree'].replace('대학교(4년)↑','대학교졸업(4년)')
```

```python
df['degree'] = df['degree'].replace('대학원 박사졸업↑','대학원 박사졸업')
df['degree'].value_counts()
```
